#  Final Game

## Please Read Before you Play!!

When running our game, PLEASE make sure this is the only thing running on your desktop. Memory leaks and RAM overuse are very common if it not. The game also crashes occasionally. We attribute all of these complications to the fact that we have over 2500 if/elif/else statements in the game. We recommend having at minimum 2G of RAM to run the game. (4/8 is preffered and 16 is required to run and edit smoothely)

#### A few items to note. For any input presented to the user, the choices are presented in the parenthesis. For example, If you want to start a piece, the pieces that are availible to start are presented as (1/2/3/4) and you must choose between those four. Failure to do so will reprompt for an input and the game will continue as before.

#### To begin the game, type Trouble() and put the desired number of players in the parenthesis. The only availible options are 2,3 and 4 players. Failure to put 2,3, or 4 in the parenthesis will either not run the program if no number is entered, or reprompt if an incorrect number is entered.
